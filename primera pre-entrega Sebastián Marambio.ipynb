Creamos una función que agregue un nuevo usuario a la base de datos:

```python
def registro (bd):
  while True:
    try:
      nuevo_usuario = {}
      nuevo_usuario["usuario"] = input("Ingresa un nombre usuario: ")
      for user in bd:
        if nuevo_usuario["usuario"] == user["usuario"]:
          raise Exception("Lo siento ese nombre de usuario ya existe")
      nuevo_usuario["pass"] = input("Ingresa una contraseña: ")
      bd.append(nuevo_usuario)
    except Exception as x:
      print(str(x))
    else:
      guardar_bd(bd)
      print("Usuario registrado exitosamente")
      break
```

Generar un string para mostrar de manera ordenada la los usuarios y su contraseña:

```python
def mostrar_bd(bd):
  info = ""
  for user in bd:
    nombre = user["usuario"]
    password = user["pass"]
    info += f"Nombre de usuario: {nombre} Contraseña: {password} \n"
  print(info)
```

Función que verifica existencia del usuario y si coincide con su respectiva clave:

```python
def login(bd):
  leer_bd(bd)
  usuario = input("Ingrese su nombre de usuario: ")
  password = input("Ingrese su contraseña: ")
  while True:
    try:
      existe = []
      for user in bd:
        if user["usuario"] == usuario:
          existe.append(user)
          if user["pass"] != password:
            raise Exception("Contraseña incorrecta")
      if len(existe) == 0:
        raise Exception("Nombre de usuario no existe")
    except Exception as x:
      print(str(x))
      break
    else:
      print("Se inicio sesión exitosamente")
      break
```

Función que guarda en un archivo txt la base de datos para su uso posterior y una que lo lee:

```python
import json
def guardar_bd(bd):
  try:
    archivo = open(ruta + "data.json", "w+")
    archivo.write(json.dumps(bd))
    archivo.close()
  except:
    print("Algo a salido mal")
  else:
    print("Archivo creado o actualizado")

def leer_bd(bd):
   archivo = open(ruta + "data.json", "r")
   contenido = archivo.read()
   archivo.close()
   return json.loads(contenido)
```
